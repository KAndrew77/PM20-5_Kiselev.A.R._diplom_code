```python
model = GraphicalLasso(alpha=0.25)
model.fit(final)

precision_matrix = model.precision_
partial_corr_matrix = -precision_matrix / np.sqrt(np.outer(np.diag(precision_matrix), np.diag(precision_matrix)))
np.fill_diagonal(partial_corr_matrix, 1)

G = nx.Graph()

for i, col in enumerate(final.columns):
    G.add_node(i, label=col)

for i in range(len(partial_corr_matrix)):
    for j in range(i + 1, len(partial_corr_matrix)):
        if abs(partial_corr_matrix[i, j]) > 0.01:
            G.add_edge(i, j, weight=partial_corr_matrix[i, j])

pos = nx.spring_layout(G, seed=42)

node_sizes = [1000 for _ in G.nodes()] 
node_colors = plt.cm.viridis(np.linspace(0, 1, len(G.nodes())))

plt.figure(figsize=(15, 15))  

nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, alpha=0.9)


edges = G.edges(data=True)
edge_colors = [edge[2]['weight'] for edge in edges]
edge_cmap = plt.cm.coolwarm
nx.draw_networkx_edges(G, pos, edgelist=[(edge[0], edge[1]) for edge in edges], width=2, edge_color=edge_colors, edge_cmap=edge_cmap, alpha=0.7)

labels = {i: G.nodes[i]['label'] for i in G.nodes}
nx.draw_networkx_labels(G, pos, labels, font_size=12, font_color='black', font_weight='bold')
cmap = plt.get_cmap('viridis')

plt.title('Частичная корреляционная сеть', fontsize=20)
plt.axis('off')  # Отключаем оси
plt.show()
```

```python
out_degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

centrality_df = pd.DataFrame({
    'OutDegree': out_degree_centrality.values(),
    'Closeness': closeness_centrality.values(),
    'Betweenness': betweenness_centrality.values()
})

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(centrality_df['OutDegree'], bins=10, kde=False)
plt.title('Прочность узла (максимальная)')
plt.xlabel('Сила')
plt.ylabel('Количество')
plt.grid(True)

plt.subplot(1, 3, 2)
sns.histplot(centrality_df['Closeness'], bins=20, kde=False)
plt.title('Близость и центральность')
plt.xlabel('Близость')
plt.ylabel('Количество')
plt.grid(True)

plt.subplot(1, 3, 3)
sns.histplot(centrality_df['Betweenness'], bins=20, kde=False)
plt.title('Промежуточность и центральность')
plt.xlabel('Промежуточность')
plt.ylabel('Количество')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.covariance import GraphicalLasso
from sklearn.cluster import KMeans

model = GraphicalLasso(alpha=0.25)
model.fit(final)

precision_matrix = model.precision_

partial_corr_matrix = -precision_matrix / np.sqrt(np.outer(np.diag(precision_matrix), np.diag(precision_matrix)))
np.fill_diagonal(partial_corr_matrix, 1)

G = nx.Graph()

for i, col in enumerate(final.columns):
    G.add_node(i, label=col)

for i in range(len(partial_corr_matrix)):
    for j in range(i + 1, len(partial_corr_matrix)):
        weight = partial_corr_matrix[i, j]
        if abs(weight) > 0.01:
            G.add_edge(i, j, weight=weight)

pos = nx.spring_layout(G)
pos_array = np.array([pos[i] for i in sorted(G.nodes())])

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(pos_array)
labels = kmeans.labels_

cmap = plt.get_cmap('viridis')
colors = [cmap(label / n_clusters) for label in labels]

nx.draw_networkx_nodes(G, pos, node_size=500, node_color=colors)

edges = G.edges(data=True)
for edge in edges:
    if edge[2]['weight'] > 0:
        color = '#003399'
    else:
        color = '#FF9933'
    nx.draw_networkx_edges(G, pos, edgelist=[(edge[0], edge[1])], width=2, edge_color=color)

labels_dict = {i: G.nodes[i]['label'] for i in G.nodes}
nx.draw_networkx_labels(G, pos, labels_dict, font_size=10)

plt.title('Частичная корреляционная сеть с кластеризацией по K-средним значениям')
plt.show()
```

```python
silhouette_avg = silhouette_score(pos_array, labels)

davies_bouldin_avg = davies_bouldin_score(pos_array, labels)

calinski_harabasz_avg = calinski_harabasz_score(pos_array, labels)

print("Silhouette Coefficient:", silhouette_avg)
print("Davies-Bouldin Index:", davies_bouldin_avg)
print("Calinski-Harabasz Index:", calinski_harabasz_avg)
```

```python
X_transposed = final.iloc[:, :-1].T

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_transposed)

input_dim = X_scaled.shape[1]
encoding_dim = 2  

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation='relu')(input_layer)
decoder = Dense(input_dim, activation='sigmoid')(encoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

autoencoder.fit(X_scaled, X_scaled, epochs=50, batch_size=32, shuffle=True, validation_split=0.2)

encoder_model = Model(inputs=input_layer, outputs=encoder)
X_encoded = encoder_model.predict(X_scaled)

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
kmeans.fit(X_encoded)
feature_labels = kmeans.labels_

G = nx.Graph()
for i, col in enumerate(final.iloc[:, :-1].columns):
    G.add_node(i, label=col)

partial_corr_matrix = np.corrcoef(final.iloc[:, :-1].T)

for i in range(len(partial_corr_matrix)):
    for j in range(i + 1, len(partial_corr_matrix)):
        if abs(partial_corr_matrix[i, j]) > 0.01:
            G.add_edge(i, j, weight=partial_corr_matrix[i, j])

pos = nx.spring_layout(G)
cmap = plt.get_cmap('viridis')
node_colors = [cmap(label / n_clusters) for label in feature_labels]

nx.draw_networkx_nodes(G, pos, node_size=500, node_color=node_colors)

# Рисование ребер
edges = G.edges(data=True)
for edge in edges:
    if edge[2]['weight'] > 0:
        color = '#003399'
    else:
        color = '#FF9933'
    nx.draw_networkx_edges(G, pos, edgelist=[(edge[0], edge[1])], width=2, edge_color=color)

labels_dict = {i: G.nodes[i]['label'] for i in G.nodes}
nx.draw_networkx_labels(G, pos, labels_dict, font_size=10)

plt.title('Partial Correlation Network with Autoencoder and K-means Clustering')
plt.show()
```

```python
silhouette_avg = silhouette_score(X_encoded, feature_labels)

davies_bouldin_avg = davies_bouldin_score(X_encoded, feature_labels)

calinski_harabasz_avg = calinski_harabasz_score(X_encoded, feature_labels)

print("Silhouette Coefficient:", silhouette_avg)
print("Davies-Bouldin Index:", davies_bouldin_avg)
print("Calinski-Harabasz Index:", calinski_harabasz_avg)
```
